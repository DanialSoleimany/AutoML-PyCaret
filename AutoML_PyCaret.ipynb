### Load a sample dataset from PyCaret

```python
from pycaret.datasets import get_data
```

### Heart Disease dataset for classification

```python
data = get_data('heart_disease')
```

### Import classification module from PyCaret

```python
from pycaret.classification import *
```

### Initialize the setup

```python
clf1 = setup(data=data, target='Disease', session_id=123)
```

### Compare multiple classification models automatically

```python
best_model = compare_models()
```

### Finalize the best model for deployment or prediction

```python
final_model = finalize_model(best_model)
```

### Launch an interactive evaluation dashboard (Confusion Matrix, AUC, etc.)

```python
evaluate_model(final_model)
```

### Make predictions on unseen/test data

```python
unseen_data = data.sample(20)
predictions = predict_model(final_model, data=unseen_data)
predictions
```
